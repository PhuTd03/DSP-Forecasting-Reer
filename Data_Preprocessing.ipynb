# GDP
Using squeeze=True to acquire the data as a Series instead of Dataframe.

```python
file_GBP = "raw_data/gdp_GBP.csv"
gdpGBP = pd.read_csv(file_GBP, index_col="DATE", parse_dates=True)
gdpGBP = gdpGBP.squeeze(True)

#chage the name
gdpGBP.name = "GDP in GBP"

gdpGBP
```

```python
file_USD = "raw_data/gdp_USA.csv"
gdpUSD = pd.read_csv(file_USD, index_col="DATE", parse_dates=True)
gdpUSD = gdpUSD.squeeze(True)

#chage the name
gdpUSD.name = "GDP in USD"

gdpUSD
```

```python
# concate the two Series and merge with the forex data
gdp = pd.concat([gdpGBP, gdpUSD], axis=1, keys=['gdpGBP', 'gdpUSD'])
gdp.dtypes
```

```python
# Convert the index of gdp to datetime
gdp.index = pd.to_datetime(gdp.index)

# Convert the index of forex to the same type as gdp
forex.index = forex.index.astype(gdp.index.dtype)
```

```python
# merging the two dataframes forex and gdp on the index
data = pd.merge(forex, gdp, left_index=True, right_index=True, how='outer')
data

```

# Current Account

```python
file_name = "raw_data/CA.csv"
data_ca = pd.read_csv(file_name)
data_ca.drop(columns={"SUBJECT", "Subject", "Country", "MEASURE", "Measure", "FREQUENCY", "Frequency", "Time", "Unit Code", "Unit", "PowerCode Code", "PowerCode", "Reference Period Code", "Reference Period", "Flag Codes", "Flags"}, inplace=True)

data_ca
```

```python
# df of the data for USA
data_ca_USA = data_ca[data_ca["LOCATION"] == "USA"]
data_ca_USA.drop(columns={"LOCATION"}, inplace=True)
data_ca_USA.set_index("TIME", inplace=True)
data_ca_USA
```

```python
# df of the data for GBR
data_ca_GBP = data_ca[data_ca["LOCATION"] == "GBR"]
data_ca_GBP.drop(columns={"LOCATION"}, inplace=True)
data_ca_GBP.set_index("TIME", inplace=True)
data_ca_GBP
```

```python
# merge the USA and GBR data, each quarter
data_ca = pd.merge(data_ca_GBP, data_ca_USA, left_index=True, right_index=True, how='outer', suffixes=('_GBP', '_USA'))

# rename the columns TIME to Date
data_ca.index.names = ['Date']

# to csv
data_ca.to_csv('raw_data/data_ca.csv')
```

```python
data_ca.index
```

```python
# Convert the index of data to datetime
data_ca.index = pd.to_datetime(data_ca.index.str.replace(r'(Q\d)-(\d+)', r'\2-\1'))
data_ca
```

```python
# Indexes became datetime object
data_ca.index
```

```python
# merge dataset
data = pd.merge(data, data_ca, left_index=True, right_index=True, how='outer')
data
```

# Libor rate

```python
file_name = "raw_data/libor_data_USD.csv"
liborUSD = pd.read_csv(file_name, index_col="Date", parse_dates=True)
liborUSD = liborUSD.squeeze(True)
# drop all columns except Avg
liborUSD = liborUSD["Average"]
liborUSD
```

```python
file_name = "raw_data/libor_data_GBP.csv"
liborGBP = pd.read_csv(file_name, index_col="Date", parse_dates=True)
liborGBP = liborGBP.squeeze(True)
# drop all columns except Avg
liborGBP = liborGBP["Average"]
liborGBP
```

```python
# concate the two Series and and make them dataframe
libor = pd.concat([liborUSD, liborGBP], axis=1, keys=['liborUSD', 'liborGBP'])
libor
```

```python
# Some of the values are in "-"
libor[libor["liborUSD"] == "-"]

# # Convert Series values to numeric
# libor = libor.apply(pd.to_numeric, errors='coerce')
# libor.info()
```

```python
dataset = pd.merge(data, libor, left_index=True, right_index=True, how='outer')
dataset[:5]
```

```python
# interpolation needs continuous datetime
# or numerical objects for index values
dataset.reset_index(level=0, inplace=True)
dataset
```

```python
# deleta all rows with NaN values after 2024-06-15
dataset = dataset.dropna(subset=['AUD'])
dataset
```

```python
# fill the missing values by interpolation
dataset.iloc[:, -6:] = dataset.iloc[:, -6:].interpolate()
dataset
```

```python
# drop rows where there is no data for GBP (or any forex)
dataset = dataset[dataset.GBP.notna()]
dataset
```

```python
# there is not enough data for libor series at the beginning
# I drop the beginning rows of the whole dataframe
dataset = dataset[dataset.liborGBP.notna()]
dataset
```

```python
# count the number of NaN values in the value_GDP
dataset.isna().sum()
```

```python
# examine the final data
rename_dict = {
    'index': 'Date',
    'Value_GBP': 'GBR_Value',
    'Value_USA': 'USA_Value',
    'liborUSD': 'liborUSD',
    'liborGBP': 'liborGBP'
}

# Renaming the columns
dataset.rename(columns=rename_dict, inplace=True)
dataset.info()
dataset.describe()
```

```python
# write dataset to dataset.csv
dataset.to_csv('dataset.csv', index=False)
```
